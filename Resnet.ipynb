### val_accuracy 기준 best model 학습

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, WeightedRandomSampler
from torchvision import datasets, transforms, models
from sklearn.metrics import precision_score, recall_score, f1_score
from collections import Counter
import mlflow
import numpy as np

## 데이터 경로 및 데이터 증강 설정
train_dir = "./Resnet_ng/train"  # 훈련 데이터 경로
val_dir = "./Resnet_ng/val"  # 검증 데이터 경로
train_transforms = transforms.Compose([
    transforms.RandomRotation(10),                  # 최대 10도 회전
    transforms.RandomHorizontalFlip(p=0.5),         # 좌우 반전
    transforms.RandomVerticalFlip(p=0.5),           # 상하 반전
    transforms.ColorJitter(brightness=0.4,          # 밝기 조정
                           contrast=0.4,
                           saturation=0.4,
                           hue=0.02),               # 색조 조정
    transforms.RandomAffine(degrees=0,              # Affine 변환
                            translate=(0.1, 0.1),
                            scale=(0.9, 1.1)),
    transforms.ToTensor(),                          # 텐서 변환
    transforms.Normalize(mean=[0.485, 0.456, 0.406],  # ImageNet 사전 학습 평균/표준편차
                         std=[0.229, 0.224, 0.225])
])
val_transforms = transforms.Compose([
    transforms.ToTensor(),  # 텐서 변환
    transforms.Normalize(mean=[0.485, 0.456, 0.406], 
                         std=[0.229, 0.224, 0.225])  # 정규화 (ImageNet 통계 사용)
])

## 데이터셋 로드
train_dataset = datasets.ImageFolder(root=train_dir, transform=train_transforms)
val_dataset = datasets.ImageFolder(root=val_dir, transform=val_transforms)

## 클래스 불균형 해결 (WeightedRandomSampler)
class_counts = Counter([label for _, label in train_dataset.samples])
class_weights = {cls: 1.0 / count for cls, count in class_counts.items()}
sample_weights = [class_weights[label] for _, label in train_dataset.samples]
sampler = WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)

## DataLoader 정의
train_loader = DataLoader(train_dataset, batch_size=16, sampler=sampler, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=4)

## 모델 정의 (ResNet-18)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet18(pretrained=True)  # ResNet-18 사용
model.fc = nn.Linear(model.fc.in_features, 4)  # 출력 클래스 수를 4로 변경
model = model.to(device)

## 손실 함수 및 옵티마이저
criterion = nn.CrossEntropyLoss()  # 가중치를 주려면 nn.CrossEntropyLoss(weight=...) 추가 가능
optimizer = optim.Adam(model.parameters(), lr=0.001)

## MLflow 설정
if mlflow.active_run() is not None:
    mlflow.end_run()
with mlflow.start_run() as run:
    mlflow.set_tag("mlflow.runName", "ResNet18_with_Aug")  # MLflow 실행 이름 설정

    ## 학습 루프
    epochs = 25  # 학습 반복 횟수 설정
    best_accuracy = 0.0
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0
        
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            correct_train += predicted.eq(labels).sum().item()
            total_train += labels.size(0)
            
        train_accuracy = correct_train / total_train
        mlflow.log_metric("Train_Loss", running_loss / len(train_loader), step=epoch)  # Train Loss
        mlflow.log_metric("Train_Accuracy", train_accuracy, step=epoch)  # Train Accuracy
        print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader):.4f}, Train Accuracy: {train_accuracy:.4f}")

        ## 검증 단계
        model.eval()
        correct_val = 0
        total_val = 0
        all_labels = []
        all_preds = []
        with torch.no_grad():  # 그래디언트 비활성화
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = outputs.max(1)
                correct_val += predicted.eq(labels).sum().item()
                total_val += labels.size(0)
                all_labels.extend(labels.cpu().numpy())  # 정답 레이블 저장
                all_preds.extend(predicted.cpu().numpy())  # 예측값 저장
                
        val_accuracy = correct_val / total_val  # Validation Accuracy 계산

        # Precision, Recall, F1-Score 계산
        precision = precision_score(all_labels, all_preds, average='macro')  # 매크로 평균
        recall = recall_score(all_labels, all_preds, average='macro')  # 매크로 평균
        f1 = f1_score(all_labels, all_preds, average='macro')  # 매크로 평균
        mlflow.log_metric("Val_Accuracy", val_accuracy, step=epoch)  # Validation Accuracy
        mlflow.log_metric("Val_Precision", precision, step=epoch)  # Precision
        mlflow.log_metric("Val_Recall", recall, step=epoch)  # Recall
        mlflow.log_metric("Val_F1", f1, step=epoch)  # F1-Score
        print(f"Validation Accuracy: {val_accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")

        ## 최적 모델 저장
        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            torch.save(model.state_dict(), "best_resnet18.pth")
            mlflow.log_artifact("best_resnet18.pth")  # MLflow에 모델 파일 저장
            print("Best model saved.")
    print("Training complete!")
    ## 최종 Hyperparameter 및 Best Accuracy 로그
    mlflow.log_param("Batch_Size", 16)
    mlflow.log_param("Learning_Rate", 0.001)
    mlflow.log_param("Epochs", epochs)
    mlflow.log_metric("Best_Val_Accuracy", best_accuracy)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, WeightedRandomSampler
from torchvision import datasets, transforms, models
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
from collections import Counter
import mlflow
import mlflow.pytorch
from torchvision.transforms.functional import InterpolationMode

# 데이터 경로 및 클래스 수
train_dir = "./Resnet_ngok/train"
val_dir = "./Resnet_ngok/val"
num_classes = 5  # OK + NG_1 ~ NG_4

# 데이터 증강 (NG 클래스 증강 강도 증가)
train_transforms = transforms.Compose([
    transforms.RandomRotation(20, interpolation=InterpolationMode.BILINEAR),  # NG 데이터에 유리한 회전 강도 증가
    transforms.RandomHorizontalFlip(p=0.7),  # NG 데이터 증강 강화
    transforms.RandomVerticalFlip(p=0.5),
    transforms.ColorJitter(brightness=0.6, contrast=0.6, saturation=0.6, hue=0.03),
    transforms.RandomAffine(degrees=10, translate=(0.2, 0.2), scale=(0.8, 1.2), shear=10),
    transforms.RandomResizedCrop(size=224, scale=(0.8, 1.0)),  # 랜덤 크롭 증강
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
val_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# 데이터셋 로드
train_dataset = datasets.ImageFolder(root=train_dir, transform=train_transforms)
val_dataset = datasets.ImageFolder(root=val_dir, transform=val_transforms)

# 클래스 불균형 해결 (WeightedRandomSampler)
class_counts = Counter([label for _, label in train_dataset.samples])
class_weights = {cls: 1.0 / count for cls, count in class_counts.items()}
class_weights_tensor = torch.tensor([class_weights[i] for i in range(num_classes)], dtype=torch.float)
sample_weights = [class_weights[label] for _, label in train_dataset.samples]
sampler = WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)

# DataLoader 정의
train_loader = DataLoader(train_dataset, batch_size=8, sampler=sampler, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False, num_workers=4)

# 모델 정의 (EfficientNet-B0)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.efficientnet_b0(pretrained=True)
model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
model = model.to(device)

# 손실 함수 및 옵티마이저
criterion = nn.CrossEntropyLoss(weight=class_weights_tensor.to(device))
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)  # 학습률 감소 스케줄러

# MLflow 설정
if mlflow.active_run() is not None:
    mlflow.end_run()

with mlflow.start_run() as run:
    mlflow.set_tag("mlflow.runName", "EfficientNetB0_ngok_recall")

    # 학습 루프
    epochs = 30
    best_recall = 0.0

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            correct_train += predicted.eq(labels).sum().item()
            total_train += labels.size(0)

        train_accuracy = correct_train / total_train
        scheduler.step()  # 학습률 업데이트
        mlflow.log_metric("Train_Loss", running_loss / len(train_loader), step=epoch)
        mlflow.log_metric("Train_Accuracy", train_accuracy, step=epoch)

        # 검증 단계
        model.eval()
        correct_val = 0
        total_val = 0
        all_labels = []
        all_preds = []

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = outputs.max(1)
                correct_val += predicted.eq(labels).sum().item()
                total_val += labels.size(0)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(predicted.cpu().numpy())

        val_accuracy = correct_val / total_val
        precision = precision_score(all_labels, all_preds, average='macro', zero_division=0)
        recall = recall_score(all_labels, all_preds, average='macro', zero_division=0)
        f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)

        mlflow.log_metric("Val_Accuracy", val_accuracy, step=epoch)
        mlflow.log_metric("Val_Precision", precision, step=epoch)
        mlflow.log_metric("Val_Recall", recall, step=epoch)
        mlflow.log_metric("Val_F1", f1, step=epoch)

        print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader):.4f}, "
              f"Train Accuracy: {train_accuracy:.4f}, Val Accuracy: {val_accuracy:.4f}, "
              f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}")

        # Recall 기준 최적 모델 저장
        if recall > best_recall:
            best_recall = recall
            torch.save(model.state_dict(), "best_efficientnet_ngok_recall.pth")
            mlflow.pytorch.log_model(model, "model")
            print("Best model saved based on Recall.")

    mlflow.log_param("Batch_Size", 8)
    mlflow.log_param("Learning_Rate", 0.001)
    mlflow.log_param("Epochs", epochs)
    mlflow.log_metric("Best_Recall", best_recall)

print("Training complete!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, WeightedRandomSampler
from torchvision import datasets, transforms, models
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, confusion_matrix
from collections import Counter
import mlflow
import mlflow.pytorch

# Early Stopping 클래스 정의
class EarlyStopping:
    def __init__(self, patience=10, delta=0, verbose=False, path='best_model.pth'):
        self.patience = patience
        self.delta = delta
        self.verbose = verbose
        self.best_score = None
        self.counter = 0
        self.early_stop = False
        self.path = path

    def __call__(self, val_recall, model):
        score = val_recall

        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_recall, model)
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.verbose:
                print(f"EarlyStopping counter: {self.counter} out of {self.patience}")
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_recall, model)
            self.counter = 0

    def save_checkpoint(self, val_recall, model):
        torch.save(model.state_dict(), self.path)
        if self.verbose:
            print(f"Validation Recall improved. Model saved to {self.path}")

# 데이터 경로 및 클래스 수
train_dir = "./Resnet_ngok/train"
val_dir = "./Resnet_ngok/val"
num_classes = 5  # OK + NG_1 ~ NG_4

# 데이터 증강
train_transforms = transforms.Compose([
    transforms.RandomRotation(10),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.02),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
val_transforms = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# 데이터셋 로드
train_dataset = datasets.ImageFolder(root=train_dir, transform=train_transforms)
val_dataset = datasets.ImageFolder(root=val_dir, transform=val_transforms)

# 클래스 불균형 해결 (WeightedRandomSampler)
class_counts = Counter([label for _, label in train_dataset.samples])
class_weights = {cls: 1.0 / count for cls, count in class_counts.items()}
sample_weights = [class_weights[label] for _, label in train_dataset.samples]
sampler = WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)

# DataLoader 정의 (배치 크기 통일)
batch_size = 16
train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=sampler, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=4)

# 모델 정의 (ResNet-18)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet18(pretrained=True)
model.fc = nn.Linear(model.fc.in_features, num_classes)
model = model.to(device)

# 손실 함수 및 옵티마이저
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10, eta_min=1e-5)

# Early Stopping 초기화
early_stopping = EarlyStopping(patience=10, verbose=True, path="best_resnet_ngok_recall_6.pth")

# MLflow 설정
if mlflow.active_run() is not None:
    mlflow.end_run()

with mlflow.start_run() as run:
    mlflow.set_tag("mlflow.runName", "ResNet18_ngok_recall6")

    # 데이터 및 모델 정보 기록
    mlflow.log_param("Dataset Train Size", len(train_dataset))
    mlflow.log_param("Dataset Validation Size", len(val_dataset))
    mlflow.log_param("Classes", ", ".join(train_dataset.classes))
    mlflow.log_param("Model Architecture", "ResNet-18")

    # 하이퍼파라미터 및 Best Recall 기록 초기화
    best_recall = 0.0
    mlflow.log_param("Batch_Size", batch_size)
    mlflow.log_param("Learning_Rate", 0.001)
    mlflow.log_param("Epochs", 80)

    # 학습 루프
    epochs = 80
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            correct_train += predicted.eq(labels).sum().item()
            total_train += labels.size(0)

        train_accuracy = correct_train / total_train
        scheduler.step()
        mlflow.log_metric("Train_Loss", running_loss / len(train_loader), step=epoch)
        mlflow.log_metric("Train_Accuracy", train_accuracy, step=epoch)
        print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader):.4f}, Train Accuracy: {train_accuracy:.4f}")

        # 검증 단계
        model.eval()
        correct_val = 0
        total_val = 0
        all_labels = []
        all_preds = []

        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = outputs.max(1)
                correct_val += predicted.eq(labels).sum().item()
                total_val += labels.size(0)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(predicted.cpu().numpy())

        val_accuracy = correct_val / total_val
        precision = precision_score(all_labels, all_preds, average='macro', zero_division=0)
        recall = recall_score(all_labels, all_preds, average='macro', zero_division=0)
        f1 = f1_score(all_labels, all_preds, average='macro', zero_division=0)

        mlflow.log_metric("Val_Accuracy", val_accuracy, step=epoch)
        mlflow.log_metric("Val_Precision", precision, step=epoch)
        mlflow.log_metric("Val_Recall", recall, step=epoch)
        mlflow.log_metric("Val_F1", f1, step=epoch)
        print(f"Validation Accuracy: {val_accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}")

        # Recall 기준 최적 모델 저장
        if recall > best_recall:
            best_recall = recall
            torch.save(model.state_dict(), "best_resnet_ngok_recall6.pth")
            mlflow.pytorch.log_model(model, "model")
            print("Best model saved based on Recall.")

        # Early Stopping 체크
        early_stopping(recall, model)
        if early_stopping.early_stop:
            print("Early stopping triggered.")
            break

    # 최종 Hyperparameter 및 Best Recall 로그
    mlflow.log_metric("Best_Recall", best_recall)
```
